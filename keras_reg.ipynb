# Keras Regression Model

Jupyter notebook for building a Regression model in Keras

Import libraries

```
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split

import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
from keras.layers.convolutional import Conv2D 
from keras.layers.convolutional import MaxPooling2D 
from keras.layers import Flatten 
```

Load the data about concrete compressive strength 

```
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
concrete_data.head()
```

We check the shape of the dataset and wether there are missing values

```
concrete_data.shape
```

```
concrete_data.describe()
```

```
concrete_data.isnull().sum()
```

The data looks clean so we split them into predictors and targets

```
concrete_data_columns = concrete_data.columns

predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
target = concrete_data['Strength'] # Strength column
```

```
predictors.head()
```

```
target.head()
```

Now we split the data in training and test sets

```
X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.30, random_state=42)
```

## Part A

Building a baseline regression model

```
n_cols = X_train.shape[1] # number of predictors

# define regression model
def regression_model():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error',metrics=['mean_squared_error'])
    return model
```

Now we built and fit the model

```
model = regression_model()
model.fit(X_train, y_train, epochs=50, verbose=2)
```

We evaluate the model on the test data:

```
scores = model.evaluate(X_test, y_test, verbose=0)
```

```
scores[1]
```

Now we repeate this calculation 50 times

```
mse_list=[]

for i in range(50):

  X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.30)

  model = regression_model()
  model.fit(X_train, y_train, epochs=50, verbose = 2)

  scores = model.evaluate(X_test, y_test, verbose=0)

  mse_list.append(scores[1])
```

And we calculate the mean and stdv of the mean squared errors.

```
len(mse_list)
```

```
print("Mean: {} \n Stdv: {}".format(np.mean(mse_list), np.std(mse_list)))
```

## Part B

Repeat the analysis with standardized data

```
predictors_norm = (predictors - predictors.mean()) / predictors.std()
predictors_norm.head()
```

```
mse_list=[]

for i in range(50):

  X_train, X_test, y_train, y_test = train_test_split(predictors_norm, target, test_size=0.30)

  model = regression_model()
  model.fit(X_train, y_train, epochs=50, verbose = 2)

  scores = model.evaluate(X_test, y_test, verbose=0)

  mse_list.append(scores[1])
```

```
len(mse_list)
```

```
print("Mean: {} \n Stdv: {}".format(np.mean(mse_list), np.std(mse_list)))
```

The standarzation improves the stdv significantly but the mean is still large

## Part C

Repeat Part B but with 100 epochs for training

```
mse_list=[]

for i in range(50):

  X_train, X_test, y_train, y_test = train_test_split(predictors_norm, target, test_size=0.30)

  model = regression_model()
  model.fit(X_train, y_train, epochs=100, verbose = 2)

  scores = model.evaluate(X_test, y_test, verbose=0)

  mse_list.append(scores[1])
```

```
len(mse_list)
```

```
print("Mean: {} \n Stdv: {}".format(np.mean(mse_list), np.std(mse_list)))
```

The extra epochs definetely helped to reduce bothe the mean and stdv

## Part D

Adding extra layers



```
# define regression model
def regression_model_v2():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error',metrics=['mean_squared_error'])
    return model
```

Train the model 50 times

```
mse_list=[]

for i in range(50):

  X_train, X_test, y_train, y_test = train_test_split(predictors_norm, target, test_size=0.30)

  model = regression_model_v2()
  model.fit(X_train, y_train, epochs=50, verbose = 2)

  scores = model.evaluate(X_test, y_test, verbose=0)

  mse_list.append(scores[1])
```

```
len(mse_list)
```

```
print("Mean: {} \n Stdv: {}".format(np.mean(mse_list), np.std(mse_list)))
```

The 3 hidden layers model produces the best performance with the smallest mean for the mean squared errors.
